```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import catboost as cb
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
```

```python
df= pd.read_csv('/content/sample_data/california_housing_train.csv')
df = df.iloc[:1598,[3,4,5,6]]
```

```python
df
```

```python
x = np.array(df['population']).reshape(-1,1)
y = np.array(df['households']).reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
clf = XGBRegressor(random_state = 1)
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)*100
```

```python
clf.predict(X_train)
```

```python
rd = RandomForestRegressor(random_state = 1)
rd.fit(X_train,y_train)
```

```python
rd.score(X_train,y_train)*100
```

```python
rd.predict(X_train)
```

```python
rgd = GradientBoostingRegressor(random_state = 1)
rgd.fit(X_train,y_train)
```

```python
rgd.score(X_train,y_train)*100
```

```python
ada = AdaBoostRegressor()
ada.fit(X_train,y_train)
```

```python
ada.score(X_train,y_train)*100
```

CatBoost

```python
c = cb.CatBoostRegressor(random_state = 1)
c.fit(X_train,y_train)
```

```python
c.score(X_train,y_train)*100
```

```python
c.predict(X_train)
```

```python
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test)
```

LIGHT GBM

```python
light = lgb.train({}, train_data, num_boost_round=100, valid_sets=[test_data])
```

```python
light.predict(X_train)
light.predict(X_test)
```
